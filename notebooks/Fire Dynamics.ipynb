```python
import numpy as np
```

```python
Q=300
W_o=0.8
H_o=2.0
L=3.6
W=2.4
H=2.4
T_inf=20
delta=0.016
k=0.48
rho=1440
c_p=840
t=300
```

```python
k = k / 1000
```

```python
hrrs = np.arange(Q + 1)
```

```python
# Calculate thermal diffusivity
alpha = k / (rho * c_p)
```

```python
# Calculate thermal penetration time
t_p = delta**2 / (4 * alpha)
```

```python
# Calculate ventilation area (m2)
A_o = W_o * H_o
```

```python
# Calculate effective heat transfer coefficient
if t < t_p:
    h_k = np.sqrt(k * rho * c_p / t)
elif t >= t_p:
    h_k = (k / delta)
```

```python
A_T = (2 * L * W) + (2 * L * H) + (2 * W * H) - (A_o)
```

```python
Delta_T = 6.85 * (hrrs**2 / (A_o * np.sqrt(H_o) * h_k * A_T))**(1/3)
```

```python
T_g = Delta_T + T_inf
```
